## Representation of Music in ScoreDraft
- Note Sequence (for instrument)
    - __Original: seq_inst=[(1.0, 48), (1.25, 48), (1.5, 48)]__
    - import ScoreDraftNotes.py： seq_inst=[do(5,48), mi(5,48), so(5,48)]
- Beat Sequence (for percussion)
    - __Original: seq_perc=[(0,48),(1,48),(0,48),(1,48)]__
    - with assist-funcs: seq_perc= [dong(48), cha(48), dong(48), cha(48)]
- Singing Sequence
    - __seq_sing=[("あ", do(5,48), "い", re(5,48), "う", mi(5,48)), BL(48),...]__
- Rap Sequence
    - __Original: seq_rap=[("jin", 48,1.0,1.0, "ke",48,1.0, 1.0, "la",48, 1.0, 0.5)]__
    - import ScoreDraftRapChinese.py：
        seq_rap=[CRap（"jin",1,48)+CRap（"ke",1,48)+CRap（"la",4,48)]

## Comparision: GUI vs. Script Interface

- Advantages of GUI
    - Intuitive
    - Easy for time-alignment
    - Easy to adjust parameters
- Adantages of Script Interface
     - Flexibility: multiple instruments can share the same track
     - Unlimitedly extendable: multi-sequence types, multi-engines\
     - Programatic sound generation
     - Light-weightness
     - Easier to input with keyboard than with mouse?
- You decide it!
